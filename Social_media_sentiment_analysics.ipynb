```python
#Import librares

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
```

```python
#Load dataframe

df = pd.read_csv(r"C:\Users\HARDIK\Documents\Techno_hacks\Social media sentiment analysis\Tweets.csv")
```

```python
#select the necessary columns

df = df[['text', 'airline_sentiment']]
```

```python
#Split the data into text and labels

X = df['text'] 
y=  df['airline_sentiment']
```

```python
# vectrorize the text data

vectorizer = TfidfVectorizer()
X_vectors = vectorizer.fit_transform(X)
```

```python
#train the data

classifier = LinearSVC()
classifier.fit(X_vectors, y)
```

```python
#Predict the data

y_pred = classifier.predict(X_vectors)
df['predicted_sentiment'] = y_pred
```
